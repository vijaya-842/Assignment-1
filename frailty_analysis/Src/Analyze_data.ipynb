```python
from google.colab import files
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
# Step 1: Upload the file
uploaded = files.upload()

# Step 2: Load the data into a DataFrame
df = pd.read_csv('cleaned_frailty_data.csv')

# Display the first few rows
print(df.head())
```

```python
# Summary statistics for the dataset (e.g., mean, min, max, etc.)
summary = df.describe()
print("Summary statistics:\n", summary)

# Compute the correlation matrix to identify relationships between numerical features
correlation_matrix = df.corr()
print("\nCorrelation matrix:\n", correlation_matrix)

# Scatter plot showing the relationship between Age and Grip Strength
sns.scatterplot(x='Age (years)', y='Grip Strength (kg)', data=df)
plt.title('Grip Strength vs Age')
plt.show()

# Box plot to compare Grip Strength between Frailty (Y/N) categories
sns.boxplot(x='Frailty (Y/N)', y='Grip Strength (kg)', data=df)
plt.title('Grip Strength by Frailty')
plt.show()

# Histogram to visualize the distribution of Grip Strength
df['Grip Strength (kg)'].hist(bins=10)
plt.title('Distribution of Grip Strength')
plt.xlabel('Grip Strength (kg)')
plt.ylabel('Frequency')
plt.show()

# Bar plot showing the average Grip Strength for each Frailty status
df.groupby('Frailty (Y/N)')['Grip Strength (kg)'].mean().plot(kind='bar')
plt.title('Average Grip Strength by Frailty Status')
plt.ylabel('Average Grip Strength (kg)')
plt.show()

# Heatmap to visualize correlations between numerical features in the dataset
sns.heatmap(df.corr(), annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap')
plt.show()
```
